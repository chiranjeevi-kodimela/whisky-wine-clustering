```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load preprocessed data
whisky = pd.read_csv("data/whiskies.txt")
regions = pd.read_csv("data/regions.txt", header=None)
whisky["Region"] = regions
flavors = whisky.iloc[:, 2:14]

# Correlation of flavors (flavor  flavor)
corr_flavors = flavors.corr()

# Correlation of whiskies (whisky  whisky)
corr_whisky = pd.DataFrame.corr(flavors.transpose())

# Plot flavor correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_flavors, cmap="coolwarm", annot=True)
plt.title("Flavor-to-Flavor Correlation")
plt.tight_layout()
plt.savefig("outputs/flavor_correlation_matrix.png")
plt.show()

# Plot whisky correlation heatmap
plt.figure(figsize=(10, 8))
plt.pcolor(corr_whisky)
plt.colorbar()
plt.title("Whisky-to-Whisky Correlation")
plt.axis("tight")
plt.tight_layout()
plt.savefig("outputs/whisky_correlation_matrix.png")
plt.show()
```
